# craete dataframes by using list concept

```python
# step 1:
# for making of dataframes 
# here the basic syntax is 
pd.DataFrame()
```

```python
# for example create dataframe by using of below mentioned list
name=['swamy','swamy1','swamy2'],
age=[20,23,25]
name,age
```

```python
pd.DataFrame(name,age)
```

```python
# provide columns in dataset
# here we can use the zip 
data=zip(name,age)
cols=['Name','Age']
pd.DataFrame(data,columns=cols)
```

```python
# provide index in dataset
data=zip(name,age)
cols=['Name','Age']
ind=['A','B','C']
pd.DataFrame(data,columns=cols,index=ind)
```

```python
# if you want to add a new column
# df['new column']
# you need to have a list which is
# having some elements that elements
# need to equal to number of rows
# city_names=['Hyd','Blr','Chennai']
# df['city']=city_names
```

```python
# provide a new column
data=zip(name,age)
cols=['Name','Age']
ind=['A','B','C']
df=pd.DataFrame(data,columns=cols,index=ind)
city_names=['xyz','xyz1','xyz2']
df['city']=city_names
df
```

```python
# update the column
data=zip(name,age)
cols=['Name','Age']
ind=['A','B','C']
df=pd.DataFrame(data,columns=cols,index=ind)
city_names=['xyz','xyz1','xyz2']
df['city']=city_names
df['Age']=[21,22,23]
df
```

```python
#In order to drop the column
#We need to use drop method
#It takes 3 parameters
#drop column or row
#mention the column name
#axis
#axis=1 reference as column
#axis=0 reference as row
#you want to create a new dataframe or
#you want overwrite the existing dataframe
#inplace= True
```

```python
# drop method
data=zip(name,age)
cols=['Name','Age']
ind=['A','B','C']
df=pd.DataFrame(data,columns=cols,index=ind)
city_names=['xyz','xyz1','xyz2']
df['city']=city_names
df['Age']=[21,22,23]
df.drop('city',axis=1,inplace=True)
df
```

```python
# to avoid the above problems
# give index=False
df.to_csv("output.csv",index=False)
pd.read_csv("output.csv")
```
